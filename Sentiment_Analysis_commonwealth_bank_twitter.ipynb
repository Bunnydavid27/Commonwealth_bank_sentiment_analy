```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Data/cleaned_commonwealth_bank_tweet_data.csv', index_col=0)
```

```python
df
```

```python
df['text']
```

```python
Analyzer = SentimentIntensityAnalyzer()
```

```python
def polarity_scores(x):
    polarity_score = Analyzer.polarity_scores(x)
    return polarity_score['compound']
```

```python
df['polarity_scores'] = df['text'].apply(polarity_scores)
df
```

```python
def posneg(x):
    if x>0:
        y = 'positive'
    elif x<0:
        y = 'negative'
    else:
        y = 'neutral'
    return y
```

```python
df['sentiment'] = df['polarity_scores'].apply(posneg)
df
```

```python
import re
def hashtags_extract(x):
    pattern = r'\#\w+'
    hashtags = re.findall(pattern, x)
    return hashtags
```

```python
df['hashtags'] = df['text'].apply(hashtags_extract)
df
```

```python
sentiments = df.sentiment.value_counts()
sentiments
```
