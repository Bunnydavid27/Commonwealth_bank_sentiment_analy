```python
from ntscraper import Nitter
import pandas as pd
```

```python
dataset_link = 'https://twitter.com/CommBank/'
```

```python
scraper = Nitter(log_level=1, skip_instance_check=False)
```

```python
Nitter.get_profile_info
```

```python
tweets = scraper.get_tweets('CommBank', mode='user', number=100)
```

```python
tweets
```

```python
len(tweets['tweets'])
```

```python
Commbank_tweets = tweets['tweets']
Commbank_tweets
```

```python
common_wealthbank_df = pd.DataFrame(Commbank_tweets)
```
