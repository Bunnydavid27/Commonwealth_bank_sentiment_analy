```python
def engagement_comments(x):
    comments = ''.join(x).split(',')[0]
    return comments
```

```python
def engagement_retweets(x):
    retweets = ''.join(x).split(',')[1]
    return retweets
```

```python
def engagement_quotes(x):
    quotes = ''.join(x).split(',')[2]
    return quotes
```

```python
def engagement_likes(x):
    likes = ''.join(x).split(',')[3]
    return likes
```

```python
df['comments']= df['stats'].apply(engagement_comments)
df['retweets']= df['stats'].apply(engagement_retweets)
df['quotes']= df['stats'].apply(engagement_quotes)
df['likes']= df['stats'].apply(engagement_likes)
```
