```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder 
```

```python
df = pd.read_csv('Data/commonwealthbank_twitter_data.csv', index_col=0)
df
```

```python
df['link'][0]
```

```python
df['text'][0]
```

```python
df = df.drop(['replying-to', 'user', 'gifs'], axis=1)
df
```

```python
df['is-retweet'].value_counts()
```

```python
df['is-pinned'].value_counts()
```

```python
df['external-link'].isna().sum()
```

```python
df['external-link'] = df['external-link'].fillna('No_External_Link')
df
```

```python
df['quoted-post'] = df['quoted-post'].replace('{}','No_Quoted_post')
```

```python
df['pictures'] = df['pictures'].replace('[]','No_Pictures')
df['videos'] = df['videos'].replace('[]','No_Videos')
df
```

```python
df['stats'][0][2:-1]
```

```python
def stats_clean(x):
    x = x[2:-1]
    return x
```

```python
df['stats'] = df['stats'].apply(stats_clean)
df
```

```python
le_retweet= LabelEncoder()
le_pinned= LabelEncoder()
```

```python
df['is-retweet'] = le_retweet.fit_transform(df['is-retweet'])
df['is-pinned'] = le_pinned.fit_transform(df['is-pinned'])
```

```python
df
```

```python
df.isna().sum()
```

```python
df.to_csv('Data/cleaned_commonwealth_bank_tweet_data.csv')
```
